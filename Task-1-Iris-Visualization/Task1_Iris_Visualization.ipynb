# Task 1: Exploring and Visualizing Iris Dataset

## Objective
The objective of this project is to explore and visualize the Iris dataset using Python libraries such as Pandas, Matplotlib, and Seaborn.

## Tools & Libraries
- Python
- Pandas
- Matplotlib
- Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset from CSV file
iris = pd.read_csv('/iris.csv')

# Display first 5 rows
iris.head()
```

```python
# Shape of dataset
print("Shape of Dataset:")
print(iris.shape)

# Column names
print("\nColumn Names:")
print(iris.columns)

# First 5 rows
print("\nFirst 5 Rows:")
print(iris.head())
```

```python
# Dataset information
print(iris.info())

# Statistical summary
print("\nStatistical Summary:")
print(iris.describe())
```

```python
# Scatter Plot
plt.figure(figsize=(8,6))

sns.scatterplot(
    x='sepal_length',
    y='petal_length',
    hue='species',
    data=iris
)

plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")

plt.show()
```

```python
# Histogram
plt.figure(figsize=(8,6))

plt.hist(iris['sepal_length'], bins=20)

plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")

plt.show()
```

```python
# Box Plot
plt.figure(figsize=(8,6))

sns.boxplot(
    x='species',
    y='petal_length',
    data=iris
)

plt.title("Petal Length by Species")

plt.show()
```

# Conclusion

In this task, the Iris dataset was explored and visualized using Python libraries such as Pandas, Matplotlib, and Seaborn.

Key findings:
